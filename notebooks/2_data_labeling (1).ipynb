```python
import pandas as pd
import os


csv_path = "/content/drive/MyDrive/10Acadamy/data/cleaned_messages.csv"
output_path = "/content/drive/MyDrive/10Acadamy/amharic_e-commerce_data_extractor/data/labeled_data.txt"

df = pd.read_csv(csv_path)
texts = df['clean_text'].dropna().sample(10, random_state=42)

lines = []


for text in texts:
  tokens = text.strip().split()
  for token in tokens:
    lines.append(f"{token} 0")
  lines.append("")

with open(output_path, 'w', encoding='utf-8') as f:
  f.write('\n'.join(lines))

print(f"✅ Sample 10 messages tokenized and saved to: {output_path}")
```

##Script to Convert CoNLL to Hugging Face Format

```python
from datasets import Dataset
import pandas as pd

# Path to your labeled file
file_path = "/content/drive/MyDrive/10Acadamy/amharic_e-commerce_data_extractor/data/labeled_data.txt"

# Load and parse the data
tokens = []
labels = []
sent_tokens = []
sent_labels = []

with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            if sent_tokens:
                tokens.append(sent_tokens)
                labels.append(sent_labels)
                sent_tokens = []
                sent_labels = []
        else:
            parts = line.split()
            if len(parts) >= 2:
                token = parts[0]
                label = parts[1]
                sent_tokens.append(token)
                sent_labels.append(label)

# Handle last sentence
if sent_tokens:
    tokens.append(sent_tokens)
    labels.append(sent_labels)

# Create Hugging Face dataset
hf_dataset = Dataset.from_dict({
    "tokens": tokens,
    "ner_tags": labels
})

# Preview the result
hf_dataset[0]
```
